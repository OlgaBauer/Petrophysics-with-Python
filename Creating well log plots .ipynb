### Importing libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import lasio
```

```python
las = lasio.read(r"O:\WSC\IWT\Python\OB\LAS files\4231331231.las")
```

```python
df = las.df()
df.head()
```

```python
df.reset_index(inplace=True)
```

```python
df.plot('GR', 'DEPT')
```

### Creatring a subplot

```python
df.plot(subplots=True, figsize=(15,15))
```

### Working with Subplots in Matplotlib

```python
fig = plt.subplots(figsize=(10,10))

#Set up a plot axis
ax1 = plt.subplot2grid((1,1), (0,0), rowspan=1, colspan=1)

ax1.plot("GR", "DEPT", data = df, color = "green") # Call the data from the df dataframe
ax1.set_xlabel("Gamma") # Assign a track title
ax1.set_xlim(0, 200) # Change the limits for the curve being plotted
ax1.set_ylim(4700,3500) # Set the depth range
ax1.grid() # Display the grid
```

### Adding a resistivity log

```python
fig = plt.subplots(figsize=(10,10))

#Set up a plot axis
ax1 = plt.subplot2grid((1,2), (0,0), rowspan=1, colspan=1)
ax2 = plt.subplot2grid((1,2),(0,1), rowspan=1, colspan=1)

ax1.plot("GR", "DEPT", data = df, color = "green") # Call the data from the df dataframe
ax1.set_xlabel("Gamma") # Assign a track title
ax1.set_xlim(0, 200) # Change the limits for the curve being plotted
ax1.set_ylim(4700,3500) # Set the depth range
ax1.grid() # Display the grid

ax2.plot("RT","DEPT", data = df, color = "red")
ax2.set_xlabel("True Res.")
ax2.set_xlim(0.2, 2000)
ax2.semilogx() # Change x axis to logariphmic scale
ax2.set_ylim(4700, 3500)
ax2.grid()
```

### Adding a density log

```python
fig = plt.subplots(figsize=(10,10))

#Set up the plot axes
ax1 = plt.subplot2grid((1,3), (0,0), rowspan=1, colspan = 1) 
ax2 = plt.subplot2grid((1,3), (0,1), rowspan=1, colspan = 1)
ax3 = plt.subplot2grid((1,3), (0,2), rowspan=1, colspan = 1)

ax1.plot("GR", "DEPT", data = df, color = "green") # Call the data from the well dataframe
ax1.set_xlabel("Gamma") # Assign a track title
ax1.set_xlim(0, 200) # Change the limits for the curve being plotted
ax1.set_ylim(4700, 3500) # Set the depth range
ax1.grid() # Display the grid

ax2.plot("RT","DEPT", data = df, color = "red")
ax2.set_xlabel("True Res.")
ax2.set_xlim(0.2, 2000)
ax2.semilogx() # Change x axis to logariphmic scale
ax2.set_ylim(4700, 3500)
ax2.grid()

ax3.plot("RHOB", "DEPT", data = df, color = "red")
ax3.set_xlabel("Bulk Density")
ax3.set_xlim(1.95, 2.95)
ax3.set_ylim(4700, 3500)
ax3.grid()
```

### Tidying up Common Elements

```python
fig, axes = plt.subplots(figsize=(10,10))

curve_names = ['Gamma', 'True Res', 'Bulk Density']


#Set up the plot axes
ax1 = plt.subplot2grid((1,3), (0,0), rowspan=1, colspan = 1) 
ax2 = plt.subplot2grid((1,3), (0,1), rowspan=1, colspan = 1)
ax3 = plt.subplot2grid((1,3), (0,2), rowspan=1, colspan = 1)



#Set up the individual log tracks / subplots
ax1.plot("GR", "DEPT", data = df, color = "green")
ax1.set_xlim(0, 200) 

ax2.plot("RT", "DEPT", data = df, color = "red")
ax2.set_xlim(0.2, 2000)
ax2.semilogx()

ax3.plot("RHOB", "DEPT", data = df, color = "red")
ax3.set_xlim(1.95, 2.95)

#Set up common elements between subplots
for i, ax in enumerate(fig.axes):
    ax.set_ylim(4700,3500)
    ax.grid()
    ax.set_xlabel(curve_names[i])
```

```python
fig, axes = plt.subplots(figsize=(10,10))

curve_names = ['Gamma', 'True Res', 'Bulk Density']


#Set up the plot axes
ax1 = plt.subplot2grid((1,3), (0,0), rowspan=1, colspan = 1) 
ax2 = plt.subplot2grid((1,3), (0,1), rowspan=1, colspan = 1)
ax3 = plt.subplot2grid((1,3), (0,2), rowspan=1, colspan = 1)



#Set up the individual log tracks / subplots
ax1.plot("GR", "DEPT", data = df, color = "green")
ax1.set_xlim(0, 200) 

ax2.plot("RT", "DEPT", data = df, color = "red")
ax2.set_xlim(0.2, 2000)
ax2.semilogx()

ax3.plot("RHOB", "DEPT", data = df, color = "red")
ax3.set_xlim(1.95, 2.95)

#Set up common elements between subplots
for i, ax in enumerate(fig.axes):
    ax.set_ylim(4700,3500)
    ax.grid()
    ax.set_xlabel(curve_names[i])

#Hide tick labels on the y-axis
for ax in [ax2, ax3]:
    plt.setp(ax.get_yticklabels(), visible=False)
    
#Reduce the space between each subplot
fig.subplots_adjust(wspace=0.05)
```

```python
fig, axes = plt.subplots(figsize=(10,10))

curve_names = ['Gamma', 'True Res', 'Bulk Density', 'Neutron']


#Set up the plot axes
ax1 = plt.subplot2grid((1,3), (0,0), rowspan=1, colspan = 1) 
ax2 = plt.subplot2grid((1,3), (0,1), rowspan=1, colspan = 1)
ax3 = plt.subplot2grid((1,3), (0,2), rowspan=1, colspan = 1)
ax4 = ax3.twiny()


#Set up the individual log tracks / subplots
ax1.plot("GR", "DEPT", data = df, color = "green", lw = 0.5)
ax1.set_xlim(0, 200) 

ax2.plot("RT", "DEPT", data = df, color = "red", lw = 0.5)
ax2.set_xlim(0.2, 2000)
ax2.semilogx()

ax3.plot("RHOB", "DEPT", data = df, color = "red", lw = 0.5)
ax3.set_xlim(1.95, 2.95)

ax4.plot("NPHI", "DEPT", data =df, color = "blue", lw = 0.5)
ax4.set_xlim(45, -15)

#Set up common elements between subplots
for i, ax in enumerate(fig.axes):
    ax.set_ylim(4700,3500)
    
    ax.xaxis.set_ticks_position("top")
    ax.xaxis.set_label_position("top")
    ax.set_xlabel(curve_names[i])
    
    if i==3:
        ax.spines["top"].set_position(("axes", 1.08))
    else:
        ax.grid()

#Hide tick labels on the y-axis
for ax in [ax2, ax3]:
    plt.setp(ax.get_yticklabels(), visible=False)
    
#Reduce the space between each subplot
fig.subplots_adjust(wspace=0.05)
```
