## Data ploting
### Simple Plot Fills

```python
# set up figure and plot
plt.figure(figsize=(5,8))
plt.plot(df['GR'], df['DEPT'], c='black', lw=0.5)
# set up axes limits
plt.xlim(10,100)
plt.ylim(3000,2550)
```

### fill_betweenx on Both sides of the line

```python
# set up figure and plot
plt.figure(figsize=(5,8))
plt.plot(df['GR'], df['DEPT'], c='black', lw=0.5)

# add fill_betweenx to left
plt.fill_betweenx(df['DEPT'], 0, df['GR'], facecolor='green')

# add fill_betweenx to right
plt.fill_betweenx(df['DEPT'], df['GR'], 100, facecolor='yellow')

# set up axes limits
plt.xlim(10,100)
plt.ylim(3000,2550)
```

### Fill to a fixed value

```python
# set up figure and plot
plt.figure(figsize=(5,8))
plt.plot(df['GR'], df['DEPT'], c='black', lw=0.5)

# Using where argument to fill to a fixed value
plt.fill_betweenx(df['DEPT'], 55, df['GR'], where=df['GR']<=55, facecolor='yellow')
plt.fill_betweenx(df['DEPT'], df['GR'], 55, where=df['GR']>=55, facecolor='green')


# set up axes limits
plt.xlim(10,100)
plt.ylim(3000,2550)
```

### Color map fills

#### Variable Fill to the Left

```python
# Set up plot values and color_index
left_value = 0
right_value = 150
span = abs(left_value-right_value)

cmap=plt.get_cmap('hot_r')

color_index = np.arange(left_value, right_value, span/100)
```

```python
color_index
```

```python
cmap
```

```python
plt.figure(figsize=(5,8))
plt.plot(df['GR'], df['DEPT'], c='black', lw=0.5)

plt.xlim(10,150)
plt.ylim(3000,2550)

for index in sorted(color_index):
    index_value = (index-left_value)/span
    color = cmap(index_value)
#    print(f'{index}\t{index_value}\t{color}')
    plt.fill_betweenx(df['DEPT'], left_value, df['GR'], where=df['GR']>=index, color=color)
```

#### Variable fill to the right

```python
plt.figure(figsize=(5,8))
plt.plot(df['GR'], df['DEPT'], c='black', lw=0.5)

plt.xlim(10,150)
plt.ylim(3000,2550)

for index in sorted(color_index):
    index_value = (index-left_value)/span
    color = cmap(index_value)
#    print(f'{index}\t{index_value}\t{color}')
    plt.fill_betweenx(df['DEPT'], right_value, df['GR'], where=df['GR']>=index, color=color)
```
