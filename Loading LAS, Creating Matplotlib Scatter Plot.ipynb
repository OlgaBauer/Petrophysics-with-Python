```python
import pandas as pd
import matplotlib.pyplot as plt
import lasio
```

```python
las = lasio.read(r"FileLocation.las")
```

### Convert las into pandas dataframe

```python
df = las.df()
df.describe()
```

For this example we are going to be focusing on RHOB (bulk density) and NPHI (neutron porosity)

# Creating a crossplot/scatterplot 

```python
# Set up the scatter plot
plt.scatter(x='NPHI', y='RHOB', data=df)
plt.show()
```

We can set up the default plot size for scatterplots with plt.rcParams

```python
plt.rcParams['figure.figsize'] = (8, 8)
```

## Modifying our plot

```python
# Set up the scatter plot
plt.scatter(x='NPHI', y='RHOB', data=df)

# Change the X and Y ranges
plt.xlim(-5,100)
# For the Y axis we need to flip by passing values in the reverse order
plt.ylim(3,1)
plt.show
```

## Adding labels to Axes

```python
# Set up the scatter plot
plt.scatter(x='NPHI', y='RHOB', data=df)

# Change the X and Y ranges
plt.xlim(-5,100)
# For the Y axis we need to flip by passing values in the reverse order
plt.ylim(3,1)

# Add in labels for the axes
plt.ylabel('Bulk density (RHOB) - g/cc')
plt.xlabel('Neutron porosity (NPHI) - %')

plt.show
```

#### We can add 3rd variable to the plot through the use of colour. We will add c argument and it the Gamma Ray GR column from the dataframe.
#### To control the range of colours shown we need to pass in values to vmin and vmax.

```python
# Set up the scatter plot
plt.scatter(x='NPHI', y='RHOB', data=df, c='GR', vmin=0, vmax=100)

# Change the X and Y ranges
plt.xlim(-5,100)
# For the Y axis we need to flip by passing values in the reverse order
plt.ylim(3,1)

# Add in labels for the axes
plt.ylabel('Bulk density (RHOB) - g/cc')
plt.xlabel('Neutron porosity (NPHI) - %')

plt.show
```

## Changing colormap and adding color ramp

To understand what colors on the plot mean, we can add a color bar. As we are using plt.scatter which is a single figure, we can call upon plt.colorbar() and also pass in the label we want to diplay, like GR in this example.

To chage the color bar, we can set it rainbow using cmap argument in plt.scatter(). https://matplotlib.org/stable/tutorials/colors/colormaps.html

```python
# Set up the scatter plot
plt.scatter(x='NPHI', y='RHOB', data=df, c='GR', vmin=0, vmax=100, cmap='rainbow')

# Change the X and Y ranges
plt.xlim(-5,100)
# For the Y axis we need to flip by passing values in the reverse order
plt.ylim(3,1)

# Add in labels for the axes
plt.ylabel('Bulk density (RHOB) - g/cc')
plt.xlabel('Neutron porosity (NPHI) - %')

# Make colorbar show
plt.colorbar(label='Gamma Ray - API')

plt.show
```

### Adding grid lines and plot styling

https://matplotlib.org/stable/gallery/style_sheets/style_sheets_reference.html
To set up a style we can use plt.style.use('bmh'), bmh is particular style.


```python
# Set up a style to bmh
plt.style.use('bmh')

# Set up the scatter plot
plt.scatter(x='NPHI', y='RHOB', data=df, c='GR', vmin=0, vmax=100, cmap='rainbow')

# Change the X and Y ranges
plt.xlim(-5,100)
# For the Y axis we need to flip by passing values in the reverse order
plt.ylim(3,1)

# Add in labels for the axes
plt.ylabel('Bulk density (RHOB) - g/cc')
plt.xlabel('Neutron porosity (NPHI) - %')

# Make colorbar show
plt.colorbar(label='Gamma Ray - API')

plt.show
```
