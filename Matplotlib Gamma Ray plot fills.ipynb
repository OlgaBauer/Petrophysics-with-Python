```python
import pandas as pd
import matplotlib.pyplot as plt
import lasio
import numpy as np
```

```python
las1 = lasio.read(r"FileLocation.las")
las2 = lasio.read(r"FileLocation.las")
```

```python
df1 = las1.df()
df1.head()
```

```python
df1.reset_index(inplace=True)
df1.rename(columns={'DEPT':'DEPT'}, inplace=True)
df1.head()
```

```python
df2 = las2.df()
df2.head()
```

```python
df2.reset_index(inplace=True)
df2.rename(columns={'DEPT':'DEPT'}, inplace=True)
df2.head()
```

```python
### Data ploting
```

```python
fig = plt.subplots(figsize=(10,8))

#Set up the plot axes
ax1 = plt.subplot2grid((1,2), (0,0), rowspan=1, colspan = 1) 
ax2 = plt.subplot2grid((1,2), (0,1), rowspan=1, colspan = 1) 

#Passing data
ax1.plot(df1['GR'], df1['DEPT'], c='black', lw=0.5)
ax1.set_ylim(1050,600)
ax1.set_xlim(10,100)
ax2.plot(df2['GR'], df2['DEPT'], c='red', lw=0.5)
ax2.set_ylim(1050,600)
ax2.set_xlim(10,100)
```

```python
fig = plt.subplots(figsize=(10,8))

#Setting up the plot axes
ax1 = plt.subplot2grid((1,2), (0,0), rowspan=1, colspan = 1) 
ax2 = plt.subplot2grid((1,2), (0,1), rowspan=1, colspan = 1) 

#Passing data
ax1.plot(df1['GR'], df1['DEPT'], c='black', lw=0.5)
ax1.set_ylim(1050,600)
ax1.set_xlim(10,100)
ax2.plot(df2['GR'], df2['DEPT'], c='black', lw=0.5)
ax2.set_ylim(1050,600)
ax2.set_xlim(10,100)

#Adding fill
ax1.fill_betweenx(df1['DEPT'], 0, df1['GR'], facecolor='green')
ax1.fill_betweenx(df1['DEPT'], df1['GR'], 100, facecolor='yellow')

ax2.fill_betweenx(df2['DEPT'], 0, df2['GR'], facecolor='green')
ax2.fill_betweenx(df2['DEPT'], df2['GR'], 100, facecolor='yellow')
```

```python
fig = plt.subplots(figsize=(12,10))

#Setting up the plot axes
ax1 = plt.subplot2grid((1,2), (0,0), rowspan=1, colspan = 1) 
ax2 = plt.subplot2grid((1,2), (0,1), rowspan=1, colspan = 1) 

#Passing data
ax1.plot(df1['GR'], df1['DEPT'], c='black', lw=0.5)
ax1.set_ylim(1100,600)
ax1.set_xlim(10,100)
ax1.set_xlabel("Gamma Ray") # Assign a track title
ax1.set_ylabel("DEPTH") # Assign a track title
ax1.set_title("BARCS Well ID 101630") #Assign title
ax2.plot(df2['GR'], df2['DEPT'], c='black', lw=0.5)
ax2.set_ylim(1100,600)
ax2.set_xlim(10,100)
ax2.set_xlabel("Gamma Ray") # Assign a track title
ax2.set_ylabel("Depth") # Assign a track title
ax2.set_title("BARCS Well ID 101629")

#Adding stratigraphy as a line
ax1.axhline(y=709.5, color='red', lw=2)
ax1.axhline(y=940, color='red', lw=2)
ax2.axhline(y=691, color='red', lw=2)
ax2.axhline(y=963, color='red', lw=2)

#Adding text
ax1.text(67, 820, 'Sparta Formation', fontsize = 12)
ax2.text(67, 820, 'Sparta Formation', fontsize = 12)

#Adding fill and condition
ax1.fill_betweenx(df1['DEPT'], 65, df1['GR'], where=df1['GR']<=40, facecolor='yellow')
ax1.fill_betweenx(df1['DEPT'], 65, df1['GR'], where=df1['GR']>=35, facecolor='green')

ax2.fill_betweenx(df2['DEPT'], 65, df2['GR'], where=df2['GR']<=35, facecolor='yellow')
ax2.fill_betweenx(df2['DEPT'], 65, df2['GR'], where=df2['GR']>=30, facecolor='green')

for ax in [ax1, ax2]:
    ax.xaxis.set_label_position("bottom")
```
